```python
url = 'http://api.worldbank.org/v2/country/'
tail = "/indicator/NY.GDP.MKTP.CD?date=2012:2019"
url_list_api = []
for country in df1['id']:
    url_list_api.append(url+country+tail)




url_list_api
```

```python
year_list = []
country_name_list = []
country_code_list = []
GDP_list = []
list3 = []
list4 = []
list1 = []
list2 = []


def opening_api(url_list_api, x):
    uh1 = urllib.request.urlopen(url_list_api[x])
    data1 = uh1.read().decode()
    tree1 = ET.fromstring(data1)
    try:
        for a in range(8):
            country_name_list.append(tree1[0][1].text)
            country_code_list.append(tree1[0][2].text)
            year_list.append(tree1[a][3].text)
            GDP_list.append(tree1[a][4].text)
        return country_name_list, country_code_list, year_list, GDP_list
    except:
        pass
```

```python
for x in range(len(url_list_api)):
    opening_api(url_list_api, x)
```

```python
column_names = ['Country', 'ID', 'GDP', 'YEAR']
```

```python
df2 = pd.DataFrame({column_names[0]: country_name_list, column_names[1]: country_code_list, column_names[2] : GDP_list, column_names[3]:year_list})
df2.head()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import re

with open("Corruption Perceptions Index - Wikipedia.html", 'r', encoding="utf-8") as fd:
    soup = BeautifulSoup(fd)
    print(type(soup))
```

```python
tables = soup.find_all( "table")
print("There are {} tables".format(len(tables)))
```

```python
right_table = soup.find("table", {"class" : "wikitable sortable"})
print(right_table.prettify())
print(type(right_table))
```

```python
title = right_table.find_all("caption", recursive = False)[0].getText().strip()
print(title[0:-8])
```

```python
source = right_table.tbody.findAll("tr", recursive = False)[0]
source_list = [th for th in source.findAll("th")]
source_list
```

```python
headers1 = []
i = 0
for i in range(len(source_list)):
    headers1.append(source_list[i].getText().strip())
    i = i + 1
headers1 = headers1[2:10]
print(headers1)
new=[]
for year in headers1:
    new.append(re.sub(r'\[[^()]*\]', "", year))
new
```

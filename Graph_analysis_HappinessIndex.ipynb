```python
plt.figure(figsize = (15,5))
plt.plot(df1.Year, df1.Happiness_Index, 'b', label = 'EEUU')
plt.plot(df2.Year, df2.Happiness_Index, 'r', label='Canada' )
plt.legend(loc='upper right', fontsize = 12)
plt.title('Happiness Index in EEUU and Canada')
plt.xlabel('Year', fontsize = 13)
plt.ylabel('Happiness Index', fontsize = 13)
plt.grid(True)
plt.show()
```

```python
fig, axs = plt.subplots(1,2,figsize=(12,5))

axs[0].hist(df1.Happiness_Index, bins = 10, alpha = 0.4, color = 'blue', edgecolor= 'navy', density = True, label = 'EEUU')
axs[1].hist(df2.Happiness_Index, bins = 10, alpha = 0.4, color = 'crimson', edgecolor = 'red', label = 'Canada' )
plt.suptitle('Histogram Happiness Index EEUU and Canada')
fig.text(0.5, 0.04, 'Happiness Index', ha='center', va='center')
fig.text(0.06, 0.5, 'Value', ha='center', va='center', rotation='vertical')
plt.show()
```
